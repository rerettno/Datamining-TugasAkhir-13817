```python
# 1. Import Library
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# 2. Load Dataset
data = pd.read_csv('salesdaily.csv')
data = data.drop(columns=['Year','Month','Hour','Weekday Name'])
data.head()
```

```python
# 3. Cek Null
data.isnull().sum()
```

```python
# 4. Set Index Datum
basket_Medicine = data.set_index('datum')

basket_Medicine.head()
```

```python
# 5. One Hot Encoding
def hot_encode(x):
  if(x<= 0):
    return 0
  if(x> 0):
    return 1

basket_encoded = basket_Medicine.applymap(hot_encode)
basket_Medicine = basket_encoded

basket_Medicine
```

```python
# 6. Bangun Model
frq_items = apriori(basket_Medicine, min_support = 0.6, use_colnames = True)
```

```python
# 7. Bangun Aturan
rules = association_rules(frq_items, metric="lift", min_threshold = 1)
rules = rules.sort_values(['confidence','lift'], ascending = [False, False])
#print(rules.head())
rules.head(100)
```
